```python
import os,cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path  = r'E:\sudeep_work\jbm_assginment\data'
```

```python
healthy_image = os.path.join(data_path,'healthy')
cracked_image = os.path.join(data_path,'crack')
```

```python
data = pd.DataFrame(columns = ['image','label'])
print('---loading healthy images-----')
for i, image in enumerate(os.listdir(healthy_image)):
#     print(i,image)
    data.loc[len(data)] = [cv2.imread(os.path.join(healthy_image,image),0),'healthy']
print('---loading cracked images-----')
for i,image in enumerate(os.listdir(cracked_image)):
#     print(i,image)
    data.loc[len(data)] = [cv2.imread(os.path.join(cracked_image,image),0),'crack']
    
```

# Exploratory data analysis

```python
healthy_data = data[data.label =='healthy'].reset_index(drop=True)
cracked_data = data[data.label =='crack'].reset_index(drop=True)

for i in range(10):
    plt.title(healthy_data.ix[i,'label'])
    plt.imshow(healthy_data.ix[i,'image'])
    plt.show()
    print('break')
    plt.title(cracked_data.ix[i,'label'])
    plt.imshow(cracked_data.ix[i,'image'])
    plt.show()
```
